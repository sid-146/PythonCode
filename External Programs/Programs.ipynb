```python
# Program 7
# !Write a program to create a dictionary of Hindi words with values as their English  Translation. Provide user with an option to look it up

import json

class EnglishDict:

    def AddWord(self,Key:str,Value:str):
        with open("Dictionary.txt", "w") as File:
            Temp = {Key, Value}

            File.append(Temp)

        return "Added Successfully"


    def GetKeys(self):
        with open("Dictionary.txt") as File:
            Data = File.read()
        
        Data = json.loads(Data)
        
        return Data.keys()


    def Search(self,Key:str):
        with open("Dictionary.txt") as File:
            Data = File.read()
        
        Data = json.loads(Data)

        Value = Data.get(Key)

        print("{0} --> {1}".format(Key, Value))


    def PrintDict(self):
        with open("Dictionary.txt") as File:
            Data = File.read()

            Data = json.loads(Data)

            for key, Value in Data:
                print("{0} --> {1}".format(key, Value))

if __name__ =='__main__':

    Text = input("Enter Option")
    print('print -> to print dictionary\nadd -> to add word\nsearch-> to search word in dictionary')

    obj = EnglishDict()

    while Text.lower() != "quit":
        print('print -> to print dictionary\nadd -> to add word\nsearch-> to search word in dictionary')
        
        Text = input("Enter Option")
        
        #! print
        if Text.lower() == 'Print':
            obj.PrintDict()
        
        #! Add new word
        elif Text.lower() == 'add':
            Key,Value = map(int,input("Enter key and value seperated by white space: ").split(" "))
            Str = obj.AddWord(Key=Key,Value=Value)
            print(Str)

        #! Search Feature
        elif Text.lower() == 'search':
            Key = input("Enter word to search")
            obj.Search(Key=Key)
        
        #! get all keys
        elif Text.lower() == 'Get':
            Keys = obj.GetKeys()
            for i in range(Keys):
                print(i+1, Keys[i], end=" | ")
        
```

```python
# Program 8
# !8. Write a program to input 8 numbers from the user and display all the unique numbers  (once).

import random 


def unique(Number):
    Ls = []
    
    for i in Number:
        if i not in Ls:
            Ls.append(i)

    for i in Ls:
        print(i, end=" ")


ioList = []
for i in range(9):
    temp = random.randint(1,123)
    ioList.append(temp)

print(ioList)

print("Unique number in list")
unique(ioList)
```

```python
# Program 15
#! Write a program to check whether a number is prime or not.

Number = int(input("Enter NUmber"))

i = 2
while i < Number:
    if Number % i == 0:
        print(Number, " is not prime number")
        break

    i += 1
else:
    print("Number is prime number")
```

```python
# Program 16
#! 16. Find the greatest of 3 numbers using function 

def Greatest(num1, num2, num3):
    largest = 0    
    
    if (num1 >= num2) and (num1 >= num3):
        largest = num1
    
    elif (num2 >= num1) and (num2 >= num3):
        largest = num2
    
    else:
        largest = num3

    return largest

import random
a = random.randint(1,125)
b = random.randint(1,125)
c = random.randint(1,125)

Greatest(a,b,c)
```

```python
# program 18
#! Recursive function to calculate the sum of first n natural numbers.

def RecSum(Num):
    
    if Num <= 1:
        return Num
    else:
        return Num + RecSum(Num-1)

RecSum(random.randint(1,125))
```
